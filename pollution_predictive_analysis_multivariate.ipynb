```python
features=df2[['temp_avr','temp_max', 'temp_min', 'pres', 'u', 'v', 'prec', 'Budateteny']].to_numpy().tolist()
```

```python
features[0:2]
```

```python
target=df2['Budateteny'].tolist()
```

```python
target
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(features,target, test_size=0.40, random_state=123, shuffle = False)
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler(feature_range=(0,1))
x_train=scaler.fit_transform(x_train)
x_test=scaler.fit_transform(x_test)
import numpy
y_train=scaler.fit_transform(numpy.array(y_train).reshape(-1,1))
y_test=scaler.fit_transform(numpy.array(y_test).reshape(-1,1))
```

```python
import tensorflow
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.layers import LSTM
from tensorflow.keras.preprocessing.sequence import TimeseriesGenerator
win_length=500
num_features=8
train_generator = TimeseriesGenerator(x_train, y_train, length=win_length, sampling_rate=1)
test_generator = TimeseriesGenerator(x_test, y_test, length=win_length, sampling_rate=1)
```

```python
train_generator[0]
```

```python
test_generator[0]
```

```python
model=Sequential()
model.add(LSTM(40,return_sequences=True,input_shape=(500,8)))
model.add(LSTM(40,return_sequences=True))
model.add(LSTM(32,return_sequences=True))
model.add(LSTM(16))
model.add(Dense(1))
```

```python
model.summary()
```

```python
import tensorflow as tf
#early_stopping = tensorflow.keras.callbacks.EarlyStopping(monitor='val_loss',patience=2,mode='min')

model.compile(loss='mae',optimizer='adam',metrics=['mae'])

history = model.fit_generator(train_generator, epochs=150,
                    validation_data=test_generator,
                    shuffle=False)
                    #callbacks=[early_stopping])
```

Lets plot the train and test loss

```python
plt.plot(history.history['loss'],label='train')
plt.plot(history.history['val_loss'],label='test')
#plt.plot(history.history['mae'],label='error')
plt.legend()
```

```python
model.evaluate_generator(test_generator, verbose=0)
```
